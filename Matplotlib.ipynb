```python
import matplotlib.pyplot as plt    # Imports matplotlib package
plt.plot([1,2,3,4,5], [10,20,30,40,50])    # Basic graph plotting
plt.show()    # Displays the Graph
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4,5], [10,20,30,40,50])
plt.xlabel('This is x axis')    # Labels the x axis
plt.ylabel('this is y axis')    # Labels the y axis
plt.title('XYZ')    # Provides a Title
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as sty

sty.use('Solarize_Light2')    # Sets the Style

x1 = [1,3,5,7,9]
y1 =  [2,4,6,8,10]
x2 = [2,4,6,8,10]
y2 =[1,3,5,7,9]

# Plotting two plots a on same Graph
plt.plot(x1, y1)
plt.plot(x2, y2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as sty    # More styling options

sty.use('dark_background')

x1 = [1,3,5,7,9]
y1 =  [2,4,6,8,10]
x2 = [2,4,6,8,10]
y2 =[1,3,5,7,9]

plt.plot(x1, y1, label = 'First' )
plt.plot(x2, y2, label = 'Second')
plt.legend(loc= 'lower right')   # Show / Decides the Legend location

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as sty

sty.use('ggplot')

x1 = [1,3,5,7,9]
y1 =  [2,4,6,8,10]
x2 = [2,4,6,8,10]
y2 =[1,3,5,7,9]

plt.plot(x1, y1, label = 'First', linewidth = 3 )
plt.plot(x2, y2, label = 'Second', linewidth = 3 )
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as sty

sty.use('ggplot')

x1 = [1,3,5,7,9]
y1 =  [2,4,6,8,10]
x2 = [2,4,6,8,10]
y2 =[1,3,5,7,9]

plt.scatter(x1, y1, label = 'First', linewidth = 3 )    # To plot a sccater Plot
plt.scatter(x2, y2, label = 'Second', linewidth = 3 )
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as sty

sty.use('ggplot')

x1 = [1,3,5,7,9]
y1 =  [2,4,6,8,10]
x2 = [2,4,6,8,10]
y2 =[1,3,5,7,9]

plt.hist(x1,  label = 'First', linewidth = 3 )    # To Plot a Histogram
plt.hist( y2, label = 'Second', linewidth = 3 )
plt.legend()

plt.show()
```
