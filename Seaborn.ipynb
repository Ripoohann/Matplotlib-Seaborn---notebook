```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline   
```

```python
print(sns.get_dataset_names())   # Displays default datasets already present in the Seaborn package
```

```python
crash_df = sns.load_dataset('car_crashes')    # Loading the Datasets and storing it in a variable
```

```python
round(crash_df.head(20))    # Displays first 20 values of dataset
```

### Distribution Plot :-

```python
sns.distplot(crash_df['not_distracted'], kde=True, bins = 25)     # Used for single variable plot
```

### Joint Plot

```python
sns.jointplot(crash_df, x = 'speeding', y='alcohol' )     # Relation b/w two variables
```

```python
sns.jointplot(crash_df, x = 'speeding', y='alcohol', kind = 'reg' )    
```

```python
sns.jointplot(data = crash_df, x = 'speeding', y='alcohol', kind = 'kde' )    
```

```python
sns.jointplot(data = crash_df, x = 'speeding', y='alcohol' , kind = 'hex' )        # Compares 2 distributions
```

### KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')

sns.pairplot(tips_df, hue = 'sex', palette = 'Blues' )

sns.pairplot(tips_df)
```

### Styling

```python
sns.set_style( "white" ) # ( darkgrid, whitegrid, dark, ticks)

plt.figure(figsize=(10, 10))

sns.set_context('paper' , font_scale=1.5) # ( talk, poster)

sns.jointplot(crash_df, x = 'speeding', y='alcohol', kind = 'reg' )   

sns.despine(left = True, bottom = True, right = False, top = False)    # Removes the Spine of Graph from any of four directions
```

### Pair Grid

```python
iris_df = sns.load_dataset('iris')

iris_g = sns.PairGrid(iris_df, hue = 'species')
# iris_g.map_offdiag(plt.scatter)
iris_g.map_diag(plt.hist)

iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)

```

```python
iris_g = sns.PairGrid(iris_df, hue = 'species',
                      x_vars = ['sepal_length', 'sepal_width'], 
                      y_vars = ['petal_length', 'petal_width'])

iris_g.map(plt.scatter)
```

### Facet Grid

```python
tips_df
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row = 'smoker')
tips_fg.map(plt.hist, 'total_bill', bins = 8)
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue = 'smoker', height = 4, aspect=1.3,
                       col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
sns.set_style( "ticks" )
```

```python
kws = dict(s=50, linewidth=0.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height = 4, aspect = 1.3,
                        hue_order=['Yes', 'No'],
                        hue_kws=dict(marker=['^', 'v']))

tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

### Regression Plots

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = tips_df, markers = ['o', '^'],
           scatter_kws={'s' : 100, 'linewidths' : 0.5, 'edgecolor' : 'white'})
```
